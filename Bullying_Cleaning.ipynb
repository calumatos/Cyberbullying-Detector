```python
# changing the column names
bs = bs.rename(columns={
    'Bullied_on_school_property_in_past_12_months': 'bullied_school',
    'Bullied_not_on_school_property_in_past_12_months': 'not_bullied_school',
    'Cyber_bullied_in_past_12_months': 'cyber_bullied',
    'Custom_Age': 'age',
    'Sex': 'sex',
    'Physically_attacked': 'physically_attacked',
    'Physical_fighting': 'physical_fighting',
    'Felt_lonely': 'felt_lonely',
    'Close_friends': 'close_friends',
    'Miss_school_no_permission': 'miss_school',
    'Other_students_kind_and_helpful': 'students_support',
    'Parents_understand_problems': 'parents_support',
    'Most_of_the_time_or_always_felt_lonely': 'most_of_the_time_or_always_felt_lonely',
    'Missed_classes_or_school_without_permission': 'missed_classes_or_school_without_permission',
    'Were_underweight': 'underweight',
    'Were_overweight': 'overweight',
    'Were_obese': 'obese'
})
```

```python
# create a new column with "Yes" if at least one of the columns is "Yes"
bs['bullied'] = bs.apply(lambda row: 'Yes' if 'Yes' in row.values else 'No', axis=1)
```

```python
bs = bs.drop(columns = ['record', 'bullied_school', 'not_bullied_school', 'most_of_the_time_or_always_felt_lonely', 
                            'missed_classes_or_school_without_permission'])
```

```python
bs
```

```python
# categorical nominal variables
bs_nom = ['bullied', 'cyber_bullied', 'underweight', 'overweight', 'obese']
for column in bs_nom:
    bs[column] = bs[column].replace({'Yes': 1, 'No': 0})
```

```python
bs['sex'] = bs['sex'].replace({'Female': 1, 'Male': 0})
```

```python
# to extract the digits
bs['age'] = bs['age'].str.extract('(\d+)')
bs['age']
```

```python
# to extract the digits
bs['physically_attacked'] = bs['physically_attacked'].str.extract('(\d+)')
bs['physically_attacked']
```

```python
bs
```

```python
### Extract digit groups or ranges
bs['physical_fighting'] = bs['physical_fighting'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')
def process_groups(groups):
    result = []
    for group in groups:
        if group[1]: 
            result.extend(map(str, range(int(group[0]), int(group[1]) + 1)))
        else:
            result.append(group[0])
    return ', '.join(result)

bs['physical_fighting'] = bs['physical_fighting'].apply(process_groups)
bs['physical_fighting']
```

```python
bs
```

```python
### Get mean
# Extract digit groups or ranges
bs['physical_fighting'] = bs['physical_fighting'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')
def calculate_mean(groups):
    numbers = []
    for group in groups:
        if group[1]:  
            numbers.extend(range(int(group[0]), int(group[1]) + 1))
        else:
            numbers.append(int(group[0]))
    return sum(numbers) / len(numbers) if numbers else None
bs['mean_physical_fighting'] = bs['physical_fighting'].apply(calculate_mean)
```

```python
### Extract digit groups or ranges
bs['close_friends'] = bs['close_friends'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')
def process_groups(groups):
    result = []
    for group in groups:
        if group[1]: 
            result.extend(map(str, range(int(group[0]), int(group[1]) + 1)))
        else:
            result.append(group[0])
    return ', '.join(result)

bs['close_friends'] = bs['close_friends'].apply(process_groups)
```

```python
### Extract digit groups or ranges
bs['miss_school'] = bs['miss_school'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')
def process_groups(groups):
    result = []
    for group in groups:
        if group[1]:
            result.extend(map(str, range(int(group[0]), int(group[1]) + 1)))
        else:
            result.append(group[0])
    return ', '.join(result)
bs['miss_school'] = bs['miss_school'].apply(process_groups)
```

```python
### Get mean
# Extract digit groups or ranges
bs['miss_school'] = bs['miss_school'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')
def calculate_mean(groups):
    numbers = []
    for group in groups:
        if group[1]:  
            numbers.extend(range(int(group[0]), int(group[1]) + 1))
        else:
            numbers.append(int(group[0]))
    return sum(numbers) / len(numbers) if numbers else None

bs['mean_school_absence'] = bs['miss_school'].apply(calculate_mean)
```
